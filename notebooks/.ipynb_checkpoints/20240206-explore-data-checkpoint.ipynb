## post process

```python
dt = dt.loc[dt['line']!='Systemwide'].sort_values('month').reset_index(drop=True)
```

```python
lines = dt['line'].unique()
```

# Plot

## single line plot

```python
def dt_line(line, day_type):
    return dt[['month','num_on_time_trips','num_sched_trips','terminal_on_time_performance']].loc[(dt['line']==line) & (dt['day_type']==day_type)]
```

```python
dt_line('1', 1)
```

```python
dt_line('1', 1).plot('month', 'terminal_on_time_performance')
```

## multiline plot

```python
df = dt.loc[dt['day_type']==1]  # weekday
df_wkd = df.set_index([df['line'], df['month']]).filter(['terminal_on_time_performance']).unstack(0).droplevel(0,axis=1)
```

```python
df_wkd.plot()
```

```python
df = dt.loc[dt['day_type']==2]  # weekend
df_wkn = df.set_index([df['line'], df['month']]).filter(['terminal_on_time_performance']).unstack(0).droplevel(0,axis=1)
```

```python
df_wkn.plot()
```

# Analyses

```python
def desc_bxplt(df, sort_value='50%'):
    df1 = df.describe().transpose().sort_values(sort_value).transpose()
    display(df1)
    df1.loc[bxplt].boxplot(vert=False)
```

```python
pd.set_option('display.max_columns', None)
bxplt = ['min','25%','50%','75%','max']
```

## stats

```python
desc_bxplt(df_wkd)
```

```python
desc_bxplt(df_wkn)
```

## stats covid

```python
def df_date(df, date, after=True):
    if after:
        return df.loc[df.index>date]
    else:
        return df.loc[df.index<date]
```

```python
desc_bxplt(df_date(df_wkd,'2020-02'))
```

```python
desc_bxplt(df_date(df_wkn,'2020-02'))
```

## stats precovid

```python
desc_bxplt(df_date(df_wkd,'2020-03',after=False))
```

## stats postcovid

```python
desc_bxplt(df_date(df_wkd,'2022-12'))
```

```python
desc_bxplt(df_date(df_wkn,'2022-12'))
```
